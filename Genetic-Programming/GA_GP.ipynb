```python
import random 
import numpy  as np 
from GA import GenA
```

```python
pop_size = 50 
generation = 100
mutation_rate = 0.01
cities = {
    "A": (0,0),
    "B" : (1,3),
    "C" : (4,3),
    "D" : (2,2),
    "E" : (5,0)
}


Genetic_alg = GenA(cities,pop_size,generation,mutation_rate)
```

```python
best_tour, best_distance = Genetic_alg.simulate()
print("Best tour: ",best_tour)
print("Total distance: ",best_distance)
```

```python
import matplotlib.pyplot as plt 

x = [cities[city][0] for city in best_tour]
y = [cities[city][1] for city in best_tour]

plt.scatter(x,y,c='red',marker='x',label='cities')

plt.plot(x,y,linestyle='-',marker='x',markersize=5,color='green',label='tsp_tour')

for city , (x,y) in cities.items():
    plt.annotate(city,(x,y),textcoords="offset points",xytext=(0,10),ha='center') 

plt.title(best_tour)
plt.legend()
plt.grid()  
```
